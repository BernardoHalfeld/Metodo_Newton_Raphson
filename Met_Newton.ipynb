# ***Método de Newton***

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

* **Definindo a função f(x)**

```python
def f(x):
    return(x*x*x*x*x - 6.0)

# Plotando gráfico da função f(x)
x = np.arange(-2.0, 2.0, 0.1)
plt.figure()
plt.grid()
plt.plot(x, f(x))
plt.legend(['f (x)'])
plt.show()
```

* **Definindo sua derivada df(x)**

```python
def df(x):
    h = 0.000001
    return((f(x + h) - f(x))/h) #aproximação para derivada
```

```python
tolerancia = 0.000001
max_iter = 100
x0 = float(input("Entre com a aproximação inicial: "))
x = 0.0
i = 1
```

```python
while (math.fabs(f(x0)) > tolerancia):# valor da função em x0 maior que o erro
    x = x0 - f(x0)/df(x0) #fórmula do método de newton
    x0 = x 
    i = i + 1
    if(i >= max_iter): #excedeu-se o número de iterações
        print("RAIZ NÃO ENCONTRADA")
        break

if(i < max_iter):
    print("\n\nRaiz %f\nIterações: %i\nf(%f) = % f\n\n"%(x0, i, x0, f(x0)))
```
